# Data Visualization

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn library
df = sns.load_dataset('titanic')
```

> Plots Using Pandas

```python
# scatter plot using pandas
df.plot(kind='scatter', x='age', y='fare')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
# change color using hex codes
df.plot(kind='scatter', x='age', y='fare', color='#FE7F0E')
```

```python
df.head()
```

```python
# use groupby function in pandas for plotting 
df.groupby('sex')['fare'].mean().plot(kind='bar')
```

```python
# use groupby function in pandas for plotting 
df.groupby('sex')['fare'].mean().plot(kind='bar')
# using Matplotlib
plt.xlabel('Gender')
plt.ylabel('Average Fare')
```

> PLots Using seaborn

```python
# bar plot using seaborn on df
sns.barplot(data=df, x='survived', y='fare')
```

```python
sns.barplot(data=df, x='sex', y='fare', hue='class')
```

```python
sns.barplot(data=df, x='sex', y='age', hue='class')
```

```python
# box plot using seaborn
sns.boxplot(data=df, x='sex', y='age', hue='class')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='class')
```

```python
# for the checking of which the sex of that paid max fare(500)
sns.scatterplot(data=df, x='fare', y='age', hue='sex')
```

```python
# flipped
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# make scatter plot using plotly
import plotly.express as px

fig = px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title=dict(text='Mt Bruno Elevation'), autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
