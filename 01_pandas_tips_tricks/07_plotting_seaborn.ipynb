> Data(df)

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

> Types of plot?

> Grouping(hue/color) based on categorical variable

```python
# Scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

> Size of the plot

```python
# Scatter plot size of the figure size
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

> Color palettes of the plot

```python
# Change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
```

```python
# add labels &Title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
```

Saving plots

```python
# Scatter plot size of  figure size
plt.figure(figsize=(8, 4))

# Change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# add labels & Title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

# saving a plot 300dpi(best), 600dpi, 150dpi png
plt.savefig('./outputs/plots/scatter_plot_zz.png', dpi=300)

# saving a plot in vector file such as svg
plt.savefig('./outputs/plots/scatter_plot_zz.svg')
```

> Style of a plot

set theme

```python
# style of a plot
sns.set_theme(style='darkgrid')
# Scatter plot size of  figure size
plt.figure(figsize=(8, 4))

# Change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

Font Scale

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# Scatter plot size of  figure size
plt.figure(figsize=(8, 4))

# Change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

> How do you interpret

Interpretation
1. We have age of passengers at x-axis and fare on y-axis
2. The scatter plot shows the relationship between age and fare of titanic passengers.
3. We further grouped the passenger by their gender such as male (Green points) and female(Red points).
4. Two male passengers gave the highest fare in titanic, and they might be travelling first class.
5. The oldest person at titanic is also a male having 80 years of age as shown in the scatter plot. (As shown in Figure 1).

# Box plot

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1.5, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(8, 4))

# Change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.boxplot(data=df, x='sex', y='age', hue='sex', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

for hue='class'

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

change legend position and title

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Box plot of Age, based on Gender and Class Name of Passengers')

#`change legend position and title`
plt.legend(loc='upper right', title='Class Name')

```

Boxen plot

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Gender and Class Name of Passengers')

#change legend position and title
plt.legend(loc='upper right', title='Class Name')

```

Change the x-ticks names

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Gender and Class Name of Passengers')

#change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"])
```

Rotation of names

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Gender and Class Name of Passengers')

#change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"])

#rotate x ticks by 45 degree
plt.xticks(rotation=45)
```

Bar plot

```python
# bar plot using seaborn
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Gender and Class Name of Passengers')

#change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"])

#rotate x ticks by 45 degree
plt.xticks(rotation=0)
```

Add caps/error bars

```python
# bar plot using seaborn
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# Scatter plot size of  figure size
plt.figure(figsize=(12, 6))

# Change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)  # Add cap size
# add labels
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Gender and Class Name of Passengers')

#change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"])

#rotate x ticks by 45 degree
plt.xticks(rotation=0)
```
