## 15 Plotly Plots on Titanic Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## 1. Scatter Plot

```python
fig1 = px.scatter(df, x='age', y='fare', color='sex')
fig1.show()
```

## 2. Line Plot

```python
fig2 = px.line(df, x='age', y='fare', color='sex')
fig2.show()
```

```python
df_line = df.sort_values(by=['age'])
fig2 = px.line(df_line, x='age', y='fare', color='sex')
fig2.show()
```

## 3. Bar plot

```python
fig3 = px.bar(df, x='sex', y='fare', color='sex')
fig3.show()
```

## 4. Box Plot

```python
fig4 = px.box(df, x='sex', y='age', color='class')
fig4.show()
```

## 5. Violin Plot

```python
fig5 = px.violin(df, x='sex', y='age', color='class', box=True)
fig5.show()
```

## 6. Histogram

```python
fig6 = px.histogram(df, x='age', color='sex')
fig6.show()
```

## 7. Pie Chart

```python
fig7 = px.pie(df, values='age', names='sex')
fig7.show()
```

```python
df['sex'].value_counts()
```

```python
df_pie = df['sex'].value_counts().reset_index()
df_pie.columns = ['sex', 'counts']
fig7 = px.pie(df_pie, values='counts', names='sex')
fig7.show()
```

## 8. Scatter 3D Plot

```python
fig8 = px.scatter_3d(df, x='age', y='fare', z='survived', color='sex')
fig8.show()
```

## 9. Area Chart

```python
fig9 = px.area(df, x='fare', y='age', color='sex')
fig9.show()
```

```python
df_area = df.sort_values(by=['fare'])
fig9 = px.area(df_area, x='fare', y='age', color='sex')
fig9.show()
```

## 10. Bubble Chart

```python
df.head()
```

```python
fig10 = px.scatter(df, x='age', y='fare', size='pclass', color='sex')
fig10.show()
```

## 11. Sunburst Chart

```python
df_sunburst = df.groupby(['who', 'age']).size().reset_index(name='counts')
fig11 = px.sunburst(df_sunburst, path=['who', 'age'], values='counts')
fig11.show()
```

```python
df_sunburst = df.groupby(['sex', 'age']).size().reset_index(name='counts')
fig11 = px.sunburst(df_sunburst, path=['sex', 'age'], values='counts')
fig11.show()
```

## 12. Parallel Coordinate Plot

```python
df['sex_id'] = df['sex'].astype('category').cat.codes
fig12 = px.parallel_coordinates(df, color='sex_id', labels={'sex_id': 'sex'}, color_continuous_scale=px.colors.sequential.Plasma)
fig12.show()
```

```python
df['class_id'] = df['class'].astype('category').cat.codes
fig12 = px.parallel_coordinates(df, color='class_id', labels={'class_id': 'class'}, color_continuous_scale=px.colors.sequential.Plasma)
fig12.show()
```

## 13 . Density Contour Plot

```python
fig13 = px.density_contour(df, x='age', y='fare', color='sex')
fig13.show()
```

## 14. Ternary Plot

```python
fig14 = px.line_ternary(df, a='age', b='fare', c='survived', color='sex')
fig14.show()
```

## 15. Polar Chart

```python
df.head()
```

```python
fig15 = px.line_polar(df, r='age', theta='sex',line_close=True)
fig15.show()
```

```python
df_radar = df.groupby(['sex']).mean(numeric_only=True).reset_index()
fig15 = px.line_polar(df_radar, r='age', theta='sex', line_close=True)
fig15.show()
```

```python
df_radar = df.groupby(['class']).mean(numeric_only=True).reset_index()
fig15 = px.line_polar(df_radar, r='fare', theta='class', line_close=True)
fig15.show()
```
