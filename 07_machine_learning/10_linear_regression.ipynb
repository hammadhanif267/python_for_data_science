# Linear Regression

`Linear Regression` is a supervised machine learning algorithm where the predicted output is `continuous` and has a `constant slope`. It's used to predict values within a `continuous range`, (e.g. sales, price) rather than trying to classify them into categories (e.g. cat, dog). There are two main types:

- **Simple regression** - only one independent variable (e.g. sales)
- **Multiple regression** - more than one independent variables (e.g. sales, price)

## Simple Linear Regression

Simple linear regression is an approach for predicting a `response` using a single feature. It is assumed that the two variables are `linearly related`. Hence, we try to find a linear function that predicts the response value(y) as accurately as possible as a function of the feature or independent variable(x).

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the data tips from sns
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
# separate the features X and labels Y
X = df[['total_bill']]
# scalar = MinMaxScaler()
# X = scalar.fit_transform(X)
y = df['tip']
```

```python
# train test split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# call the model
model = LinearRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# take out model intercept and slop, make an equation
print(model.intercept_)
print(model.coef_)
print('y = ', model.intercept_, '+', model.coef_, '* X')
```

```python
model.predict([[5]])
```

```python
# prediction
y_pred = model.predict(X_test)
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred))
print('R2 = ', r2_score(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# plot the model and data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
plt.show()
```

# Multiple Linear Regression
`Multiple regression` is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. It is a type of regression analysis that allows for the inclusion of multiple predictor variables in the model.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# import the dataset
df = sns.load_dataset('tips')
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_bill', y='tip', hue='size', data=df)

plt.title("Tip vs Total Bill with Size as Hue")
plt.show()
```

```python
# separate the features X and labels Y
X = df[['total_bill', 'size']]
#scalar = MinMaxScaler()
#X = scalar.fit_transform(X)
y = df['tip']
```

```python
# train test split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# call the model
model = LinearRegression()
```

```python
# train the model
model.fit(X_train, y_train)
```

```python
# take out model intercept and slop, make an equation
print(model.intercept_)
print(model.coef_)
print('y = ', model.intercept_, '+', model.coef_, '* X')
```

```python
model.predict([[16, 2]])
```
