# Inverse encoding/ Inverse transform /Decoding categorical variables

Inverse encoding is the process of converting the encoded data back to its original form. This is done by using the `inverse_transform()` method. This method is available for the following encoders:

- `LabelEncoder`
- `OneHotEncoder`
- `OrdinalEncoder`

```python
# import libraries
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
# load the data
df = sns.load_dataset('titanic')
df.head()
# remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# impute missing values for age, embarked, embark_town and deck
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
# drop deck column
df.drop('deck', axis=1, inplace=True)
```

```python
# encode all categorical variables and object variables

le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
le_embark_town = LabelEncoder()
le_alive = LabelEncoder()

df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])
df['alive'] = le_alive.fit_transform(df['alive'])
df.head()
```

```python
# inverse transform the data

df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df['alive'] = le_alive.inverse_transform(df['alive'])
df.head()
```

# one hot encoding

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

# Load Titanic dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
categorical_cols = ['sex', 'embarked', 'pclass']  # pclass is numerical but treated as categorical
df[categorical_cols].head()
```

```python
# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  # drop='first' avoids dummy variable trap

# Fit and transform categorical data
encoded_data = encoder.fit_transform(df[categorical_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate with original dataset (excluding original categorical columns)
df_encoded = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

df_encoded.head()
```

```python
# Decode back to original categories
decoded_data = encoder.inverse_transform(encoded_df)

# Convert back to DataFrame
decoded_df = pd.DataFrame(decoded_data, columns=categorical_cols)

# Check if we recovered the original categorical data
decoded_df.head()
```

# Ordinal Encoding

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Load Titanic dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
categorical_cols = ['sex', 'embarked', 'pclass']  # pclass is numerical but can be treated as categorical
df[categorical_cols].head()
```

```python
# Initialize OrdinalEncoder
encoder = OrdinalEncoder()

# Fit and transform categorical data
encoded_data = encoder.fit_transform(df[categorical_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=categorical_cols)

# Concatenate with original dataset (excluding original categorical columns)
df_encoded = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

df_encoded.head()
```

```python
# Decode back to original categories
decoded_data = encoder.inverse_transform(encoded_df[categorical_cols])

# Convert back to DataFrame
decoded_df = pd.DataFrame(decoded_data, columns=categorical_cols)

# Check if we recovered the original categorical data
decoded_df.head()
```
